```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
import scipy.stats as sps
import statsmodels.api as ms
```

```python
daily_dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
N = len(daily_dates)
trow = sps.expon(scale=10).rvs(size=N)
plt.figure(figsize = (15,8))
trow = pd.Series(trow)
trow.index=daily_dates 

# поочередно зададим кривые (перевозки и скользящее среднее) с подписями и цветом
plt.plot(trow, label = 'Ряд', color = 'steelblue')
plt.plot(trow.rolling(window = 10).mean(), label = 'Скользящее среднее', color = 'orange')

plt.show()

# применяем функцию к данным о перевозках
decompose = ms.tsa.seasonal_decompose(trow)
decompose.plot()

plt.show()

plot_acf(trow)
plt.show()
```
