```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pns
import math as mt
import scipy.stats as sps
```

```python
dist = sps.uniform(1,20)
#dist = sps.norm(loc=5,scale=3)
N = 10000
MX = np.zeros(N)
for i in range(N):
    data  =  dist.rvs(size=1000)
    MX[i] =  data.mean()

plt.figure()
plt.title("Распределение мат.ожидания")
plt.hist(MX)
plt.show()

print('Стандартное отклонение', MX.std(ddof=1))

print(sps.sem(data))
print(data.std(ddof=1)/np.sqrt(len(data)))

std = sps.sem(data)
m = MX.mean()


intervals = sps.norm.interval(0.98,loc= MX.mean(),scale = sps.sem(data))
print(intervals)
print(m-3*std,m+3*std)
```
